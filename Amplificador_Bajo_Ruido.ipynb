## 1. Amplificador de bajo ruido.

El siguiente proyecto presenta el diseño de un amplificador de bajo ruido LNA (Low Noise Amplifier). 
Los LNA son muy usados en receptores de RF como en los teléfonos celulares, receptores de GPS, redes LANs (WiFi) y comunicación satelital. 
Este tipo de amplificadores, por lo general, se encuentra inmediatamente despues de las antenas en los sistemas de RF, aunque a veces puede haber un filtro pasa banda de por medio. 
Al ser los primeros amplificadores del sistema, su función es amplificar toda la señal, incluyendo el ruido, recibida por la antena, sin agregar mas ruido propio del amplificador. 
De esta manera se amplifica la señal y no se presentan cambios muy significativos en la relación señal a ruido a la salida. 
Son un componente clave en el front-end de los receptores de RF ya que como agregan poco ruido a la señal, se caracterizan por tener una cifra de ruido baja, y una ganancia buena, y como sabemos por la formula de Friis la cifra de ruido total del sistema es dominada por las primeras etapas del sistema. 

La cifra de ruido del LNA se minimiza eligiendo componentes de bajo ruido, el punto de operación y la topología del circuito. 
Los dispositivos activos mas usados en los LNA son fabrica dos usando GaAs pHEMTS o transistores bipolares usando SiGe, pero también se usan transistores de silicio para aplicaciones menos criticas o de bajo costo. 
En general los dispositivos de GaAs tienen menor cifra de ruido y mayor linealidad en comparación con los transistores de SiGe, pero los transistores fabricados usando SiGe trabajan de manera mas eficiente y son de mas fácil integración. 

La elección de la topología depende de la aplicación especifica que se le de al LNA, por ejemplo si queremos un amplificador con muy baja cifra de ruido podemos usar un emisor común, pero tenemos considerar que posible que necesite una estabilización externa, y no tendremos gran ancho de banda, a diferencia de la topología de base común, la cual es estable y además tiene mayor linealidad, pero menor ganancia.

El diseño del amplificador sintonizado de  pequeña señal de RF generalmente se basa en el requisito de una ganancia de potencia específica a una frecuencia dada.
Otros objetivos de diseño pueden incluir ancho de banda, estabilidad, aislamiento de entrada-salida y bajo rendimiento de ruido.
Después de seleccionar un tipo de circuito básico, se pueden resolver las ecuaciones de diseño aplicables.
Los circuitos se pueden clasificar de acuerdo con la retroalimentación (neutralización, unilateralización o sin retroalimentación), y la coincidencia en los terminales del transistor (las admisiones del circuito coinciden o no con las entradas y salidas de los transistores).
Se discutirá cada una de estas categorías de circuitos, incluidas las ecuaciones de diseño aplicables y las consideraciones que conducen a la selección de una configuración particular.

Las ecuaciones que figuran en este documento son aplicables a las configuraciones de emisor común, base común o colector común, utilizando el conjunto de parámetros correspondiente (parámetros de emisor común, base común o colector común).
Si bien se desarrolla principalmente el diseño de circuitos con transistores bipolares convencionales, la teoría de la red de dos puertos tiene la ventaja de ser aplicable a cualquier red activa lineal.
Por lo tanto, el mismo enfoque de diseño y ecuaciones pueden usarse con los transistores de efecto de campo [7], [9], los circuitos integrados [10] o cualquier otro dispositivo que pueda describirse como una red activa lineal de dos puertos.

Las componentes de conductancia y susceptancia expresada en términos de $R_{n}$ , $G_{u}$ , $G_{\gamma}$  y $B_{\gamma}$ quedan:

$$G_{opt}=[\frac{G_{u}+R_{u}G_{\gamma}^2}{R_{u}}]^{\frac{1}{2}}$$

$$B_{opt}=-B_{\gamma}$$

Y el valor mínimo de cifra de ruido es:

$$ F_{opt}=1 + 2R_{n}(G_{\gamma}+G_{opt})$$

Por lo tanto la cifra de ruido puede ser descrita mediante cuatro parámetros fundamentales $F_{opt}$, $G_{opt}$ , $B_{opt}$ y $R_{n}$. La cifra de ruido queda:

$$ F=F_{opt} + \frac{R_{n}}{G_{s}}[(G_{s}-G_{opt})^2+(B_{s}+B_{opt})^2]$$

Estos parámetros se pueden medir variando la admitancia de fuente y midiendo el ruido de salida. Cuando el ruido es mínimo, se obtiene $Y_{opt}$, el valor de potencia de ruido permite e $Y_{opt}$ permite obtener los demás parámetros.

Ahora veamos el caso del amplificador sintonizado.
El circuito de adaptación transforma la impedancia de generador $R_{g}$ a la necesaria a la entrada del transistor $R_{in}$. 
A la frecuencia de sintonía $f_{0}$, se anulan las componentes reactivas por lo tanto, la reactancia vista en la base del transistor $B'_{s}$ es igual a cero. 
Solo se ve una resistencia $R'_{s}$, o de forma equivalente una conductancia $G'_{s}$. 
La expresión para la cifra de ruido de un amplificador sintonizado seria la siguiente:

$$F=F_{opt} + \frac{R_{n}}{G'_{s}}[(G'_{s}-G_{opt})^2+B_{opt}^2]$$

Si seleccionamos $G'_{s}=G_{opt}$, se obtiene la figura de ruido mínima y es:

$$ F=F_{opt} + \frac{R_{n}B_{opt}^2}{G_{opt}}$$

Cada diseño será una relación de compromiso entre mínima cifra de ruido, la máxima trasferencia de energía y el $Q_{c}$ del circuito.

Estas definiciones de potencia se refieren a diferentes condiciones de circuito. Esto permite que se desarrollen varias definiciones de ganancia diferentes que se relacionan con diferentes etapas en el desarrollo de un amplificador e indican el rendimiento final que se puede lograr a partir de un amplificador. Ganancia del sistema:

### 1.4 Estabilidad

Existe el potencial para que un amplificador sea inestable y oscile. 
Generalmente no se trata de una oscilación constante con amplitud y frecuencia fijas, sino de una respuesta caótica. 
El diseño del oscilador no es tan sencillo como diseñar un amplificador inestable.
Para el análisis de estabilidad, el dispositivo activo en un amplificador lineal a menudo puede tratarse como un cuadripolo de dos puerto.

Las consideraciones de estabilidad afectan la ganancia máxima (estable) que se puede lograr y la facilidad de diseño. 
Si la ganancia máxima estable es demasiado baja, entonces es necesario seleccionar otro transistor. 
Si la ganancia especificada está cerca de la ganancia máxima estable, entonces el diseño será un desafío, especialmente para anchos de banda más amplios. Entonces el esfuerzo de diseño aumenta con transistores más simples (y por lo tanto más baratos). 

Hay muchas maneras de ver la estabilidad. 
En el dominio del tiempo la inestabilidad se manifiesta como el crecimiento de las señales a lo largo del tiempo independientemente del nivel de la señal de entrada. 
Sin embargo, es más eficiente analizar y diseñar circuitos de RF y microondas en el dominio de la frecuencia, por lo que la estabilidad debe evaluarse también en el dominio de la frecuencia. 
Las muchas técnicas de dominio de frecuencia disponibles para evaluar la estabilidad varían según el nivel de cobertura y la facilidad con la que se pueden aplicar. 
La técnica más simple y más fácil de aplicar se basa en el análisis de dos puertos, lo que lleva a métricas de estabilidad basadas en parámetros de dos puertos y al concepto de regiones (llamadas círculos) de estabilidad en un gráfico de Smith y una técnica más general es el análisis de estabilidad Nyquist. 

2.6.1 Análisis de estabilidad de dos puertos

El análisis de estabilidad debe aplicarse a los dos puertos más internos que contienen el dispositivo activo y cualquier red de retroalimentación. 
Si el puerto más interno de dos (incluyendo cualquier retroalimentación) es incondicionalmente estable, entonces el amplificador en el que está incrustado será estable. Así, existe un proceso de selección natural de manera que los transistores disponibles tienden a no sufrir inestabilidades internas.

El acoplamiento $\Gamma_S$ y $\Gamma_L$ dificulta el diseño independiente de las redes de coincidencia de entrada y salida. 
Es mucho más conveniente considerar la situación incondicionalmente estable en la que la entrada es estable sin importar cuál sea la red de coincidencia de carga y salida presente, y la salida es estable sin importar cuál sea la fuente y la red de coincidencia de entrada presente. 
Como primera etapa en el diseño, un amplificador lineal está diseñado para una estabilidad incondicional. 
El espacio de diseño es mayor si la prueba más rigurosa de estabilidad, se utiliza para determinar la estabilidad. Sin embargo, la ventaja (es decir, un espacio de diseño más grande) suele ser pequeña.

Si la fuente y la carga son pasivas,$|\Gamma_L|<1$ entonces $|\Gamma_S|<1$ y para que las oscilaciones se acumulen si

$|\Gamma_{IN}|>1$ y $|\Gamma_{OUT}|>1$

Para garantizar la estabilidad para todas las terminaciones pasivas de fuente y carga (es decir, estabilidad incondicional), entonces

$|\Gamma_{IN}|<1$ y $|\Gamma_{OUT}|<1$

Los amplificadores a menudo se realizan como etapas mediante las cuales una etapa de amplificador alimenta a otra. Esto complica el análisis de estabilidad, ya que es posible $\Gamma_S$ y $\Gamma_L$ ser más que la unidad. Si $\Gamma_S$ y $\Gamma_L$ son ambos menores que uno para múltiples etapas de amplificador, entonces se puede usar la estabilidad del amplificador que se describe aquí.

Para que los criterios de estabilidad sean utilizados en el diseño deben ser puestos en términos de los parámetros de dispersión del dispositivo activo. Existen dos criterios de estabilidad adecuados comúnmente utilizados como k-factor y $\mu-factor$, que ahora serán considerados.

El criterio de estabilidad $\mu-factor$, con estabilidad incondicional se obtiene si

$\mu = \frac{1−|S_{11}|^2}{|S22−S^∗_{11} \Delta|+|S_{21} S_{12}|}>1$

Un resultado importante es que un mayor valor de $\mu$ indica mayor estabilidad. 
El $\mu$ factor es una sola cantidad que proporciona una condición suficiente y necesaria para una estabilidad incondicional. 
Es decir, no importa qué fuente pasiva y carga se presenten (es decir, si $|\Gamma_S|≤1$ y$|\Gamma_L|≤1$) entonces el amplificador será estable si $\mu>1$.

También se define el parámetro dual como


$\mu^´ = \frac{1−|S_{22}|^2}{|S_{11}−S^∗_{22} \Delta|+|S_{21} S_{12}|}>1$

Esta es la distancia desde el centro de la gráfica Smith (es decir, el origen de la gráfica polar de los parámetros S) hasta el punto potencialmente inestable más cercano en el plano de carga de salida. 
Es decir, es la distancia más corta desde el origen hasta el círculo de estabilidad de salida. Si $\mu>1$ (indicando la posibilidad de inestabilidad de dos puertos), entonces $\mu^´>1$ también. 


Así, para determinar si un puerto de dos puertos es incondicionalmente estable, sólo es necesario considerar uno de ellos. 
Además,$\mu$ y a veces $\mu^´$ se les conoce como los factores de estabilidad geométrica de entrada y salida respectivamente, y a veces simplemente como $\mu_1$ (MU1) y $\mu_2$ (MU2), respectivamente. 
Calcular ambos es útil, ya que esto permite examinar la estabilidad relativa en la entrada y en la salida.


#### El transistor potencialmente inestable

Cuando el factor de estabilidad de Linvill del transistor determinado por la ecuación 1 es mayor que uno, el transistor es potencialmente inestable.
Ciertas combinaciones de admisiones de fuente y carga causarán oscilaciones si no se utiliza realimentación.
Al diseñar con el transistor potencialmente inestable, se deben tomar medidas para asegurar que el amplificador sea estable.
La estabilidad generalmente se logra mediante uno o ambos métodos:

1. utilizando una red de retroalimentación que reduce el parámetro $y_{12}$ a un valor que asegure la estabilidad.

2. Elegir una fuente y una combinación de admitancia de carga que proporciona estabilidad.
